# Basic Python Chatbot That Repeats User Input

```python
# This line prints a welcome message to the user
print("Welcome to my first chatbot!")

# This line explains what the chatbot does
print("This chatbot will repeat whatever you say.")

# This line asks the user for their name and stores it in the variable 'name'
name = input("What's your name? ")

# This line prints a greeting using the name the user entered
print("Nice to meet you, " + name + "!")

# This line asks the user to type something and stores it in 'stuff_to_echo'
stuff_to_echo = input("Enter something for me to repeat: ")

# This line repeats back what the user typed
print("You said: " + stuff_to_echo)
```
